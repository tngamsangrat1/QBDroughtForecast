```python
import datetime as dt
import glob
import h5py
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
import os
import pandas as pd
#import xarray as xr
```

```python
filename='SMAP_test.h5'
f = h5py.File(filename, 'r')
print(f); print('\n')
```

```python
#Now lets look at the groups within the file to access
i=0;
for key in f.keys():
    print(str(i)+ '\t'+key)
    i+=1
group_id=list(f.keys())[1]
```

```python
print('\n')
i=0
for var in list(f[group_id].keys()):
    print(str(i)+'\t'+var)
    i+=1
```

```python
print('the data has a shape of: '+str(f[group_id][list(f[group_id].keys())[0]].shape))
```

```python
var_id = list(f[group_id].keys())[22] # soil_moisture
sm_data = f[group_id][var_id][:]
sm_ds = f[group_id][var_id]

var_id = list(f[group_id].keys())[10] # soil_moisture
lati = f[group_id][var_id][:]

var_id = list(f[group_id].keys())[12] # soil_moisture
longi = f[group_id][var_id][:]

print('data are returned as '+str(type(sm_data)))
```

```python
sm_ds,sm_data,lati,longi
```

```python
df=pd.DataFrame(np.transpose([sm_data,lati,longi]),columns=['moisture', 'lati', 'longi'])
df
```

```python
moist=df.moisture.to_numpy()
np.max(moist)
```
